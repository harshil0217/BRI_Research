```python
#converted adjusted to float

def remove_space(x):
    one = x
    two = ""
    for s in one:
        if s.isdigit():
            two += s
    return two

chinese_loan["Adjusted Amount (Nominal USD)"] = chinese_loan[" Adjusted Amount (Nominal USD) "].apply(lambda x: remove_space(x))
    

chinese_loan["Adjusted Amount (Nominal USD)"] = pd.to_numeric(chinese_loan["Adjusted Amount (Nominal USD)"], downcast = 'float')
```
